```python
## Convert string to integer
value[1] = int(value[1])
value[2] = int(value[2])


#operation code
oc = ["0 ", i, j, k]
oc[1] = int(oc[1])
oc[2] = int(oc[2])
oc[3] = int(oc[3])


#memory address
adr = [9, 10, 11, 12]


## FUNCTION TO READ AND DISPLAY OPCODE ON INSTRUCTIONS
def akses_oc(instruction_pc): 
    pc = int(instruction_pc)
    if  (instruction_pc == 100 ):return ("0 ")
    elif(instruction_pc == 101 ):return (oc[1])
    elif(instruction_pc == 102 ):return (oc[2])
    elif(instruction_pc == 103 ):return (oc[3])

    
## FUNCTION TO READ AND DISPLAY ADDRESS ON INSTRUCTIONS
def akses_adr(instruction_pc):
    pc = int(instruction_pc)
    if  (instruction_pc == 100 ):return 0 
    elif(instruction_pc == 101 ):return (adr[1])
    elif(instruction_pc == 102 ):return (adr[2])     
    elif(instruction_pc == 103 ):return (adr[3])

## FUNCTION TO READ AND DISPLAY MEMORY VALUE AT THE ADDRESS ACCORDING TO THE INSTRUCTIONS
def akses_value(av): 
    if  (akses_adr(av) == 0      ):return (value[0])
    elif(akses_adr(av) == adr[1] ):return (value[1])
    elif(akses_adr(av) == adr[2] ):return (value[2]) 
    elif(akses_adr(av) == adr[3] ):return (value[3])   
    
    
    
##PC (Program Counter) FUNCTION 
def pc_next(pc):
    pc = int(pc)
    return pc + 1

def pc_before(pc):
    pc = int(pc)
    return pc - 1

##MAR (Memory Address Register) FUNCTION
def mar(m):
    if(m == 100):
        return (" 0 ")
    else:
        return m  

##Accumulator FUNCTION
# variabel accumulator --> ac
ac_value = [0, 0, 0, 0]
i = 0
for i in range(4):
    ac_value[i]  = value[i] 

def ac(ac): 
    if  (akses_adr(ac) == 0      ):return (ac_value[0])
    elif(akses_adr(ac) == adr[1] ):return (ac_value[1])
    elif(akses_adr(ac) == adr[2] ):return (ac_value[2]) 
    elif(akses_adr(ac) == adr[3] ):return (ac_value[3])     
    
def akses_ac(aac):
    if(akses_oc(aac) == 2 ):return akses_ac(pc_before(aac))
    else: return ac(aac)
    
    
##Arithmetic Logic Unit (ALU) FUNCTION
def alu(pc):
        if  (akses_oc(pc) == "0 "): return ac_value[0]
        elif(akses_oc(pc) == 1 or akses_oc(pc) == 2 ): return alu(pc_before(pc))
        elif(akses_oc(pc) == 5): return akses_ac(pc_before(pc))
        
jumlah_intruksi = 0    
def operasi_alu(alu, operasi):
    if(operasi == 0): 
        if(akses_oc(alu) == 1 or akses_oc(alu) == 2): return operasi_alu(pc_before(alu), operasi)
        else : return akses_ac(alu) 
    elif(operasi == 1 and akses_oc(alu) == 5):
        ac_value[jumlah_intruksi] = ac_value[jumlah_intruksi-1] + ac_value[jumlah_intruksi] 
        return ac_value[jumlah_intruksi]    

    
```

```python
### Langkah-Langkah Siklus Instruksi
def langkah2_siklus_fetch(no):
    if  (no == 1): return ("1. Pertama Program Counter (PC) akan menyimpan instruksi selanjutnya yang akan dieksekusi ")
    elif(no == 2): return ("2. Alamat instruksi pada PC disimpan didalam Memory Address Register (MAR) ")
    elif(no == 3): return ("3. Value yang ada pada alamat intruksi dimuat di Memory Data Register (MDR) ")
    elif(no == 4): return ("4. Instruksi pada MDR disimpan di Current Instruction Register (CIR) dan PC dinaikkan 1 untuk menyiapkan intruksi berikutnya " )
    
def langkah2_siklus_decode():
    return ("Unit kontrol menerima instruksi dari CIR dan menginterpretasi instruksi")
                            
def langkah2_siklus_execute(pc, no):
    if  (akses_oc(pc) == 1 ):
        return ("Accumulator mengambil dan menyimpan value dari MDR  ")

    elif(akses_oc(pc) == 5 ):                                
        if  (no == 1): return ("1. Value dari Accumulator sebelumnya dipindahkan ke ALU dan Accumulator menerima value dari MDR  ")
        elif(no == 2): return ("2. Value yang ada pada Accumulator diteruskan ke ALU dan ALU melakukan operasi penjumlahan ")
        elif(no == 3): return ("3. Hasil operasi penjumlahan pada ALU disimpan di Accumulator ")

    elif(akses_oc(pc) == 2 ):                      
        if  (no == 1): return ("1. Value yang ada pada Accumulator dipindahkan ke MDR ")
        elif(no == 2): return ("2. Value MDR disimpan di alamat memori sesuai dengan alamat intruksi")
        
                                     
     
### siklus instruksi (fetch, decode, execute) 
##--> input Program Counter (PC)   
# FETCH instruction     
def fetch(pc): 
    print(">>> SIKLUS FETCH INTRUKSI ",pc)
    print("*********************************")
    print(langkah2_siklus_fetch(1))   
    running(pc, akses_oc(pc_before(pc)), akses_adr(pc_before(pc)), mar(pc_before(pc)), akses_oc(pc_before(pc)), akses_adr(pc_before(pc)), akses_value(pc_before(pc)), akses_oc(pc_before(pc)), akses_adr(pc_before(pc)), akses_ac(pc_before(pc)), alu(pc_before(pc)), operasi_alu(pc_before(pc),0))
    print(langkah2_siklus_fetch(2))
    running(pc, akses_oc(pc_before(pc)), akses_adr(pc_before(pc)), mar(pc), akses_oc(pc_before(pc)), akses_adr(pc_before(pc)), akses_value(pc_before(pc)), akses_oc(pc_before(pc)), akses_adr(pc_before(pc)), akses_ac(pc_before(pc)), alu(pc_before(pc)), operasi_alu(pc_before(pc),0))
    print(langkah2_siklus_fetch(3))
    running(pc,akses_oc(pc_before(pc)), akses_adr(pc_before(pc)), mar(pc), akses_oc(pc), akses_adr(pc), akses_value(pc), akses_oc(pc_before(pc)), akses_adr(pc_before(pc)), akses_ac(pc_before(pc)), alu(pc_before(pc)), operasi_alu(pc_before(pc),0))
    print(langkah2_siklus_fetch(4))
    running(pc_next(pc), akses_oc(pc), akses_adr(pc), mar(pc), akses_oc(pc), akses_adr(pc), akses_value(pc), akses_oc(pc_before(pc)), akses_adr(pc_before(pc)), akses_ac(pc_before(pc)), alu(pc_before(pc)), operasi_alu(pc_before(pc),0))
    print("-----------------------------------------siklus fetch pada intruksi ",pc," selesai------------------------------------------\n\n")
        
        
# DECODE instruction
def decode(pc):
    print(">>> SIKLUS DECODE INTRUKSI ",pc)
    print("*********************************")
    print(langkah2_siklus_decode())
    running(pc_next(pc), akses_oc(pc), akses_adr(pc), mar(pc), akses_oc(pc), akses_adr(pc), akses_value(pc), akses_oc(pc), akses_adr(pc), akses_ac(pc_before(pc)), alu(pc_before(pc)), operasi_alu(pc_before(pc),0))
    print("-----------------------------------------siklus decode pada intruksi ",pc," selesai------------------------------------------\n\n")
  

# EXECUTE instruction
def execute(pc):
    print(">>> SIKLUS EXECUTE INTRUKSI ",pc)
    print("*********************************")
    if(akses_oc(pc) == 1):
        print(langkah2_siklus_execute(pc, 1))
        running(pc_next(pc), akses_oc(pc), akses_adr(pc), mar(pc), akses_oc(pc), akses_adr(pc), akses_value(pc), akses_oc(pc), akses_adr(pc), akses_ac(pc), alu(pc_before(pc)), operasi_alu(pc_before(pc),0))
        print("-----------------------------------------siklus execute pada intruksi ",pc," selesai------------------------------------------\n\n")
        
    elif(akses_oc(pc) == 5): 
        print(langkah2_siklus_execute(pc, 1)) 
        running(pc_next(pc), akses_oc(pc), akses_adr(pc), mar(pc), akses_oc(pc), akses_adr(pc), akses_value(pc), akses_oc(pc), akses_adr(pc), akses_ac(pc), alu(pc), operasi_alu(pc_before(pc),0)) 
        print(langkah2_siklus_execute(pc, 2))
        running(pc_next(pc), akses_oc(pc), akses_adr(pc), mar(pc), akses_oc(pc), akses_adr(pc), akses_value(pc), akses_oc(pc), akses_adr(pc), akses_ac(pc), alu(pc), operasi_alu(pc,1))
        print(langkah2_siklus_execute(pc, 3))
        running(pc_next(pc), akses_oc(pc), akses_adr(pc), mar(pc), akses_oc(pc), akses_adr(pc), akses_value(pc), akses_oc(pc), akses_adr(pc), akses_ac(pc), alu(pc), operasi_alu(pc,0))
        print("-----------------------------------------siklus execute pada intruksi ",pc," selesai------------------------------------------\n\n")
    
    elif(akses_oc(pc) == 2):
        print(langkah2_siklus_execute(pc, 1)) 
        running(pc_next(pc), akses_oc(pc), akses_adr(pc), mar(pc), akses_oc(pc), akses_adr(pc), akses_ac(pc), akses_oc(pc), akses_adr(pc), akses_ac(pc), alu(pc), operasi_alu(pc_before(pc),0))
        
        value[jumlah_intruksi] = ac_value[jumlah_intruksi-1]
        for i in range(4): ac_value[i]  = value[i]  
                
        print(langkah2_siklus_execute(pc, 2))  
        if(akses_oc(pc_before(pc)) == 5):     
            running(pc_next(pc), akses_oc(pc), akses_adr(pc), mar(pc), akses_oc(pc), akses_adr(pc), akses_value(pc), akses_oc(pc), akses_adr(pc), akses_value(pc), alu(pc), akses_value(pc))
        
        else :
            running(pc_next(pc), akses_oc(pc), akses_adr(pc), mar(pc), akses_oc(pc), akses_adr(pc), akses_ac(pc), akses_oc(pc), akses_adr(pc), akses_ac(pc), alu(pc), operasi_alu(pc_before(pc),0))
            
            
        print("-----------------------------------------siklus execute pada intruksi ",pc," selesai------------------------------------------\n\n")


def instruction_cycle(pc):
    if (akses_oc(pc) == 1):
        fetch(pc), decode(pc), execute(pc), print ("SIKLUS INSTRUKSI ",pc," SELESAI")
 
    if (akses_oc(pc) == 5):
        fetch(pc), decode(pc), execute(pc), print ("SIKLUS INSTRUKSI ",pc," SELESAI")
    
    if (akses_oc(pc) == 2):
        fetch(pc), decode(pc), execute(pc), print ("SIKLUS INSTRUKSI ",pc," SELESAI") 


```
